```python
# Increasing the brightness

import cmath
import cv2 as cv
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt

im1 = cv.imread('/content/drive/MyDrive/Images/Image20240626145240.jpg', cv.IMREAD_GRAYSCALE)
assert im1 is not None

im2 = cv.add(im1, 100)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(cv.cvtColor(im1, cv.COLOR_BGR2RGB), cmap=cm.gray, vmin=0, vmax=255)
ax[0].axis('off')
ax[1].imshow(cv.cvtColor(im2, cv.COLOR_BGR2RGB), cmap=cm.gray, vmin=0, vmax=255)
ax[1].axis('off')
```

```python
# Creating a (6, 8) grayscale image with a white pixel at (2, 3)
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = np.zeros((6,8), dtype=np.uint8)
im[2,3] = 255
fig, ax = plt.subplots(1, 1, figsize=(6, 8))
ax.imshow(im, cmap='gray', vmin=0, vmax=255)
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
# Creating a (6, 8) color image with a pink pixel at (2, 3)
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = np.zeros((6,8,3), dtype=np.uint8)
im[2,3] = (255, 215, 0)
fig, ax = plt.subplots(1, 1, figsize=(6, 8))
ax.imshow(im)
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
# Extracting the Blue Channel from an Image.
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = cv.imread('/content/drive/MyDrive/Images/flowers.jpeg')
im_blue = im.copy()
im_blue[:,:,1] = 0
im_blue[:,:,2] = 0
fig, ax = plt.subplots(1, 2, figsize=(12, 8))
ax[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
ax[1].imshow(cv.cvtColor(im_blue, cv.COLOR_BGR2RGB))
ax[1].set_title('Blue Channel')
for a in ax:
    a.axis('off')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = cv.imread('/content/drive/MyDrive/Images/flowers.jpeg')
im_hsv = cv.cvtColor(im, cv.COLOR_BGR2HSV)
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0, 0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax[0, 0].set_title('Original Image')
ax[0, 1].imshow(im_hsv[:,:,0], cmap='gray')
ax[0, 1].set_title('Hue Channel')
ax[1, 0].imshow(im_hsv[:,:,1], cmap='gray')
ax[1, 0].set_title('Saturation Channel')
ax[1, 1].imshow(im_hsv[:,:,2], cmap='gray')
ax[1, 1].set_title('Value Channel')
for a in ax.ravel():
    a.axis('off')
plt.show()
```
