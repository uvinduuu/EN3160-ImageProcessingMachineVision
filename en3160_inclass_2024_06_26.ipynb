```python
# Using OpenCV to read and display an image

import cv2 as cv
import numpy as np

im = cv.imread('/content/drive/MyDrive/Images/Image20240626145240.jpg')
assert im is not None

cv.namedWindow('Image', cv.WINDOW_NORMAL)
cv.imshow('Image', im)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Increasing the brightness

import cmath
import cv2 as cv
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt

im1 = cv.imread('/content/drive/MyDrive/Images/Image20240626145240.jpg', cv.IMREAD_GRAYSCALE)
assert im1 is not None

im2 = cv.add(im1, 100)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(cv.cvtColor(im1, cv.COLOR_BGR2RGB), cmap=cm.gray, vmin=0, vmax=255)
ax[0].axis('off')
ax[1].imshow(cv.cvtColor(im2, cv.COLOR_BGR2RGB), cmap=cm.gray, vmin=0, vmax=255)
ax[1].axis('off')
```
