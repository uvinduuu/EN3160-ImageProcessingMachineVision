```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 10
hw = 3*sigma #rule of thumb for the half-width of the filter use 3 times sigma

x = np.linspace(start=-hw, stop=hw, num=2*hw+1)
y = np.linspace(start=-hw, stop=hw, num=2*hw+1)

x, y = np.meshgrid(x, y)
'''
print(x) 
print(y) 
'''
#same value in each column, means y same x changing, same value in each row, means x same y changing, this gives us the x,y grid

G = 1/(2*np.pi*sigma**2) * np.exp(-(x**2 + y**2)/(2*sigma**2))
G=(G/np.sum(G)).round(4)

print(G)

#plot in 3-D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, G, cmap='coolwarm')
plt.show()
```

```python
# 1-D Gaussian of length 5 with sigma = 1

sigma = 1
hw = 2

X = np.linspace(start=-hw, stop=hw, num=2*hw+1)
g = 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-X**2/(2*sigma**2))
g
```

```python
# Outer product between two 1-D Gaussian vectors

G = np.outer(g, g)
G=(G/np.sum(G)).round(4)
G
```

```python
# 3x3 Sobel kernal seperated into x and y components

sobel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])    
print(sobel)

sobel_x = np.array([-1, 0, 1])
sobel_y = np.array([1, 2, 1])
sobel = np.outer(sobel_y, sobel_x)
print(sobel)
```
